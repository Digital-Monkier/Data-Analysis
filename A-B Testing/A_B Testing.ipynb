```python
hypothesis['ICE'] = ((hypothesis['impact'] * hypothesis['confidence']) / hypothesis['effort']).round(2)
display(hypothesis[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False))
```

## RICE 
- Reach — how many users will be affected by the update you want to introduce
- Impact — how strongly this update will affect the users, their experience, and their satisfaction with the product
- Confidence — how sure you are that your product will affect them in this way
- Effort — how much will it cost to test the hypothesis

```python
hypothesis['RICE'] = ((hypothesis['reach'] * hypothesis['impact'] * hypothesis['confidence']) / hypothesis['effort']).round(2)
display(hypothesis[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False))
```

```python
plt.figure(figsize=(10,6))
plt.title('Method ICE and RICE', size=15)
sns.scatterplot(x=hypothesis['ICE'], y=hypothesis['RICE'])
for line in range(0,hypothesis.shape[0]):
     plt.text(hypothesis.ICE[line]+0.2, 
              hypothesis.RICE[line], 
              hypothesis.index[line], 
              horizontalalignment='left',
              size='medium',
              color='black', 
              weight='semibold'
             )
plt.show()
```

# Aggregating Orders Dataset 
- Group user groups by date 
- Aggregate orders with the userGroups dataset
- count unique visitors, orders, and sum the revenue all per group
- Similar process to aggreagte visits per date using visits dataset


```python
datesGroups = orders[['date', 'group']].drop_duplicates()
datesGroups.head()
```

```python
ordersAggregated = datesGroups.apply(
    lambda x: orders[
        np.logical_and(
            # We are filtering orders on every date and group instance on that date in datesGroups
            orders['date'] <= x['date'], orders['group'] == x['group']
        )
    # Then we are taking the date, group, number of unique transaction and users and 
    # the total revenue for each group on that date
    ].agg(
        {
            'date': 'max',
            'group': 'max',
            'transactionId': pd.Series.nunique,
            'visitorId': pd.Series.nunique,
            'revenue': 'sum',
        }
    ),
    axis=1,
).sort_values(by=['date', 'group'])
```

```python
ordersAggregated.head()
```

```python
visitorsAggregated = datesGroups.apply(
    lambda x: visits[
                     np.logical_and(
                         visits['date'] <= x['date'], visits['group'] == x['group']
                     )
    ].agg(
        {
            'date':'max', 'group':'max', 'visits':'sum'
        }
      ), 
      axis=1
    ).sort_values(by=['date', 'group'])
```

```python
visitorsAggregated.head()
```

```python
cumulativeData = ordersAggregated.merge(
    visitorsAggregated, 
    left_on=['date', 'group'],
    right_on=['date', 'group']
  )

cumulativeData.columns = [
    'date',
    'group',
    'orders',
    'buyers',
    'revenue',
    'visitors'
    ]


cumulativeData.head()
```

```python
def avgStatCheck(sampA, sampB, alpha = 0.05):
    
    stat1, p1Norm = st.shapiro(sampA)
    stat2, p2Norm = st.shapiro(sampB)
    
    if p1Norm> alpha and p2Norm > alpha:
        statLev, pLev=st.levene(sampA, sampB)

        if pLev<alpha:
            tStat, pVal = ttest_ind(sampA, sampB, equal_var=False)
        else:
            tStat, pVal = ttest_ind(sampA, sampB, equal_var=True)

    else:
        mStat, pVal=st.mannwhitneyu(sampA, sampB, alternative ='two-sided')
    print('B to A Avg Difference:{0:.5f}'.format(sampB.mean()/sampA.mean()-1))
    print('p-Value:', pVal.round(5))
    
    if (pVal < alpha):
        print("Rejecting the null hypothesis: there is a significant difference between the groups")
    else:
        print("Failed to reject the null hypothesis: there is no significant difference between the groups")

```
